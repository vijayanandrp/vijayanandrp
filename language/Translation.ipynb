```python
from googletrans import Translator

# translator = Translator()

translator = Translator(
    service_urls=['translate.google.com',
                  'translate.google.co.in']
    )

target_language = {'ta': 'tamil', 'hi': 'hindi',  'kn': 'kannada', 'te': 'telugu', 'ml': 'malayalam'}
source = {'en': 'english'}
```

```python
words = {'greetings': ['Hi', 'Hello', 'Hey', 'How are you?',
                       'How have you been?', 'Howdy', 'Greetings',
                       'Good Morning',  'Afternoon',
                       'Good evening', 'Good night',
                       'How do you do',  'Pleased to meet you',
                       'Good to see you', 'Long time no see', 'how is your day'
                      'Dear Sir or Madam', 'did you have food', 'do you want to eat something',
                       'shall i bring water', 'how is family and kids',
                       'To Whom it May Concern',    'Alright', 'Ok',
                       'Allow me to introduce myself', 'Hello Stranger', 'Bye', 'take care'
                       'Good Day', 'Thanks', 'Thank you', 'Pleasure', 'Sorry',
                       'cool down', 'mistake', 'nothing to worry']
}
```

```python
translated_dict = {}

for dest in target_language.keys():
  for word in words['greetings']:
    translate = translator.translate(word, src='en', dest=dest)
    if word not in translated_dict:
      translated_dict[word] = []
    translated_dict[word].append((target_language[dest], translate.pronunciation ))

```

```python
# translated_dict
```

```python
revised_dict = []

for key in translated_dict.keys():
  value = dict(translated_dict[key])
  revised_dict.append({**{'english': key}, **value})
```

```python
import pandas as pd

df = pd.DataFrame(revised_dict)

df
```

```python
print(df.to_markdown(tablefmt="github", index=False))
```
